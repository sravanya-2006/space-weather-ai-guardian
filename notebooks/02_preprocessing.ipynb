```python
import pandas as pd

df = pd.read_csv("data/raw/solar_flares.csv")

print(df.shape)
df.head()
```

```python
df.columns
```

```python
df = df.drop(
    columns=[
        "note",
        "sentNotifications",
        "link"
    ]
)
```

```python
df.info()
```

```python
df["beginTime"] = pd.to_datetime(df["beginTime"])
df["peakTime"] = pd.to_datetime(df["peakTime"])
df["endTime"] = pd.to_datetime(df["endTime"])
df["submissionTime"] = pd.to_datetime(df["submissionTime"])
```

```python
df.info()
```

```python
df["duration_minutes"] = (
    df["endTime"] - df["beginTime"]
).dt.total_seconds() / 60
```

```python
df[["classType", "duration_minutes"]].head()
```

```python
df["flare_class"] = df["classType"].str[0]
```

```python
df["flare_class"].value_counts()
```

```python
risk_map = {
    "A": 0,
    "B": 0,
    "C": 1,
    "M": 2,
    "X": 3
}

df["risk_level"] = df["flare_class"].map(risk_map)
```

```python
df[["classType", "flare_class", "risk_level"]].head()
```

```python
import os

output_dir = "data/processed"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

df.to_csv(
    os.path.join(output_dir, "solar_flares_clean.csv"),
    index=False
)

print("Clean dataset saved successfully")
```
