```python
import pandas as pd
import numpy as np

df = pd.read_csv("solar_flares_clean.csv")

df.head()
```

```python
df["beginTime"] = pd.to_datetime(df["beginTime"])
df["peakTime"] = pd.to_datetime(df["peakTime"])
df["endTime"] = pd.to_datetime(df["endTime"])
df["submissionTime"] = pd.to_datetime(df["submissionTime"])
```

```python
df["hour"] = df["beginTime"].dt.hour
```

```python
df["day"] = df["beginTime"].dt.day
```

```python
df["month"] = df["beginTime"].dt.month
```

```python
df["year"] = df["beginTime"].dt.year
```

```python
df["day_of_week"] = df["beginTime"].dt.dayofweek
```

```python
df["magnitude"] = df["classType"].str[1:].astype(float)
```

```python
flare_map = {
    "C": 0,
    "M": 1,
    "X": 2
}

df["flare_class_encoded"] = df["flare_class"].map(flare_map)
```

```python
df[
    [
        "classType",
        "magnitude",
        "duration_minutes",
        "hour",
        "month",
        "flare_class_encoded",
        "risk_level"
    ]
].head()
```

```python
df.to_csv(
    "final_features.csv",
    index=False
)
```
